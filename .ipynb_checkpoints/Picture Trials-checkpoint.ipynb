```python
import numpy as np
import scipy.ndimage as ndimage
import matplotlib.pyplot as plt


def entropy_filter(image, size):
    # Convert image to grayscale
    grayscale_image = np.mean(image, axis=2)
    
    # Compute local entropy
    local_entropy = ndimage.generic_filter(grayscale_image, entropy, size=size)
    
    return local_entropy

def entropy(window):
    # Compute entropy of the window
    hist, _ = np.histogram(window, bins=256, range=(0, 255))
    hist = hist / np.sum(hist)  # Normalize histogram
    entropy = -np.sum(hist * np.log2(hist + 1e-10))  # Compute entropy
    return entropy

# Load the image
image = plt.imread('input_image.jpg')

# Apply entropy filter
filtered_image = entropy_filter(image, size=3)

# Save the result
plt.imsave('entropy_filtered_image.jpg', filtered_image, cmap='gray')
```
